```python
playlist_df = pd.DataFrame()

playlist_df['moods'] = playlist.keys()
playlist_df['links'] = playlist.values()
playlist_df['word_lists'] = word_lists.values()
playlist_df['word_lists'] = playlist_df['word_lists'].apply(lambda x: ','.join(map(str, x)))

playlist_df.head()
```

```python
# to display the name of the playlist 1st song
def link(http):
    url = http
    page = urlopen(url)
    soup = bs(page, 'html.parser')
    return soup.title.get_text()
```
